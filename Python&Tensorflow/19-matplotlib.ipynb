```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi=[10,20,30,30,30,40,50,60,70,80]
```

```python
kiloListesi=[20,40,50,55,57,60,70,78,88,90]
```

```python
numpyYasListesi=np.array(yasListesi)
numpyKiloListesi=np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"k")
plt.xlabel("yas")
plt.ylabel("kilo")
plt.title("kilonun yasa gore değişim grafiği")
plt.show()
```

```python
numpyDizisi1=np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2=numpyDizisi1**2
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,'b*-')
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2)
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
benimFigure = plt.figure()
figureBase = benimFigure.add_axes([0.2,0.2,0.6,0.9])
figureBase.plot(numpyDizisi1,numpyDizisi2,"g")
figureBase.set_xlabel("x ekseni")
figureBase.set_ylabel("y ekseni")
figureBase.set_title("başlık")
```

```python
figure2=plt.figure()

eksen1=figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2=figure2.add_axes([0.3,0.5,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("ana grafik başlık")


eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("x ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("ana grafik başlık")
```

```python
(benimfigurem,benimeksenler)=plt.subplots(nrows=1,ncols=2)

type(benimeksenler)
for eksen in benimeksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"b")
    eksen.set_xlabel("x ekseni")
plt.tight_layout()
```

```python
yeniFig=plt.figure()
yeniEksen=yeniFig.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 **2,label="numpydizisi**2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3,label="numpydizisi**3")
yeniEksen.legend(loc=6)
```

```python
##dpi
yeniFig.savefig("ornekfigure.png",dpi=200)
```

```python
(benFigure,benEksen)=plt.subplots()
benEksen.plot(numpyDizisi1,numpyDizisi2,color="#38A1B0",alpha=0.3)
benEksen.plot(numpyDizisi2,numpyDizisi1,color="#C98F97")
```

```python
(yenifigurem,yenieksenim)=plt.subplots()
yenieksenim.plot(numpyDizisi1,numpyDizisi1+2,color="blue",linewidth=8.0)
yenieksenim.plot(numpyDizisi1,numpyDizisi1+4,color="yellow",linewidth=5.0)
yenieksenim.plot(numpyDizisi1,numpyDizisi1+6,color="pink",linestyle="--")
yenieksenim.plot(numpyDizisi1,numpyDizisi1+5,color="pink",linestyle=":")
yenieksenim.plot(numpyDizisi1,numpyDizisi1+7,color="pink",linestyle="-.")

yenieksenim.plot(numpyDizisi1,numpyDizisi1+7,color="#000000",linestyle="-.",marker="o",markersize=8,markerfacecolor="red")
```
